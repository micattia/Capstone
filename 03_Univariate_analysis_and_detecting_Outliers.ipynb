```
# Another way, Creating function to calculate the z-Score and get outliers

import numpy as np
import pandas as pd
outliers= []
#df_outliers = pd.DataFrame(columns = 'y')
def detect_outlier(data_1):
    
    threshold=3
    mean_1 = np.mean(data_1)
    std_1 =np.std(data_1)
    
    
    for y in data_1:
        z_score= (y - mean_1)/std_1 
        if np.abs(z_score) > threshold:
            outliers.append(y)
    #df_outliers.append(outliers)
    return outliers
  
outlier_datapoints = detect_outlier(dfAPPL.Pos)
print(outlier_datapoints)
```
